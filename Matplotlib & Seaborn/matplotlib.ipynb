```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
X = np.linspace(0,5,11)
X
```

```python
Y = X**2
Y
```

```python
plt.plot(X,Y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot(X, Y, 'r')
plt.subplot(1, 2, 2)
plt.plot(Y, X, 'b')
```

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.plot(X, Y)
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_title('Title')
```

### Understanding `figure` and `add_axes` in Matplotlib (OOP Method)

---

#### **1. `fig = plt.figure()`**
Creates an empty figure  basically a blank canvas to draw plots on.

---

#### **2. `add_axes([left, bottom, width, height])`**
Adds an **axes object** (a plotting area) at a specific location inside the figure.

```python
fig = plt.figure()
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # main axes
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])  # inset axes
ax1.plot(X, Y)
ax2.plot(Y, X)
ax2.set_title('Smaller Plot')
ax1.set_title('Larger Plot')
```
