```python
import seaborn as sns
```

## Data
Seaborn comes with built-in data sets!

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
## HISTOGRAM
sns.histplot(tips['total_bill'],kde=True,bins=30)
```

```python
sns.histplot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.scatterplot(x='total_bill', y='tip', data = tips)
```

## Regplot shows a scatter plot PLUS a regression line (trend line).
It helps identify relationship direction (positive/negative) and strength.

```python
sns.regplot(x='total_bill', y='tip', data = tips)
```

```python
sns.regplot(x='total_bill', y='tip', data = tips, scatter=False)
```

## Pairplot creates pairwise plots between every numeric column in a dataset.
## Helps to quickly understand relationships & correlations across all features.

```python
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex') ## hue for categories
```

## rugplot

rugplots are actually a very simple concept, they just draw a dash mark for every point on a univariate distribution. They are the building block of a KDE plot:

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['tip'])
```

## kdeplot

kdeplots are [Kernel Density Estimation plots]These KDE plots replace every single observation with a Gaussian (Normal) distribution centered around that value.
