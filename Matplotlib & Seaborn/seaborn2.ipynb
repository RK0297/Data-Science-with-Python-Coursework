```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)## by default mean
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill',hue='smoker',data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', hue='sex', split=True, data=tips)## split for side by side comparisons
```

```python
sns.stripplot(x='day', y='total_bill', data=tips)
```

```python
sns.stripplot(x='day', y='total_bill', hue='sex', data=tips, jitter=True) ## jitter adds noise and separates overlaping points
```
