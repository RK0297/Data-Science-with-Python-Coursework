## Matrix Plots

 ## Heatmap
A Heatmap is a type of matrix plot that visually represents data in a matrix form, where the values of the matrix are displayed as colors.

```python
import seaborn as sns
import pandas as pd	
tips = sns.load_dataset('tips')	
tips.head()
```

```python
## now we convert categorical to numerical 
tips_corr = pd.get_dummies(tips,drop_first=True).corr()
tips_corr
```

```python
sns.heatmap(tips_corr, annot=True) ## annot for correlation values
```

## using pivot table

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
flg = flights.pivot_table(values='passengers',index='month',columns='year')
flg.head()
```

```python
sns.heatmap(flg)
```

```python
sns.clustermap(flg)
```

## FACETING
The process of splitting one large dataset into smaller groups and plotting the same visualization for each group.

```python
sns.lmplot(x="total_bill", y="tip", col="sex", data=tips)
```

```python
g = sns.FacetGrid(tips, col="day")
g.map(sns.scatterplot, "total_bill", "tip")
```
